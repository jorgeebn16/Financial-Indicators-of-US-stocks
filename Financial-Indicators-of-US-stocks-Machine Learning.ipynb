```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sb
import os
```

```python
# Read in csv
df = pd.read_csv("Resources/2018_Financial_Data.csv")
df.fillna(0)
# Drop the null columns where all values are null
df = df.dropna(axis='columns')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.describe()
```

# Select features (columns)

```python
# Set target, features and feature_names.
target = df["Unnamed: 0"]
data = df.drop(['Unnamed: 0','Sector'], axis=1)
feature_names = data.columns
data.head()
```

# automatically calculate feature importance

```python
from sklearn.ensemble import RandomForestClassifier
# Create a random forest classifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(data, target)
rf.score(data, target)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, data.columns), reverse=True)
```

# Build a new DataFrame Based on Importances

```python
df1 = pd.read_csv("Resources/2018_Financial_Data.csv", usecols = ['2019 PRICE VAR [%]', 'EPS Growth','EPS Diluted Growth','Net Income Growth', 'capitalExpenditureCoverageRatios','ROIC','Operating Cash Flow growth','Payables Turnover','operatingCashFlowSalesRatio','Free Cash Flow growth','Receivables growth','Net Cash/Marketcap','Earnings Yield','Net Debt to EBITDA','assetTurnover','pretaxProfitMargin','3Y Net Income Growth (per Share)','Shareholders Equity per Share','cashFlowCoverageRatios','EPS Diluted','Inventories', 'Cash and cash equivalents', 'SG&A Expense','Revenue','debtEquityRatio', 'Free Cash Flow per Share', 'Weighted Average Shs Out', 'Capex to Depreciation','Cash and short-term investments','priceFairValue','Operating Income Growth','cashRatio','Inventory Growth','Net cash flow / Change in cash',
'nIperEBT','Total current assets','Earnings Before Tax Margin', 'Effect of forex changes on cash','Other comprehensive income','Other Liabilities','Asset Growth','returnOnEquity', 'Operating Income', '10Y Revenue Growth (per Share)', 'Gross Profit Growth'])
df1.head()
```

```python
df2 = df1.astype(int)
df2.head()
```

```python
# Set target, features and feature_names.
target1 = df2["Revenue"]
data1 = df2.drop(['Revenue','2019 PRICE VAR [%]'], axis=1)
feature_names = data1.columns
data1.head()

```

# Create a Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data1, target1, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

```python
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)

X_train_minmax = X_minmax.transform(X_train)
X_test_minmax = X_minmax.transform(X_test)
```

# Train the Model (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier
rf1 = RandomForestClassifier(n_estimators=200)
print(rf1)
rf1.fit(X_train_minmax, y_train)
```

```python
print(f"Training Data Score: {rf1.score(X_train_minmax, y_train)}")
print(f"Testing Data Score: {rf1.score(X_test_minmax, y_test)}")
```

```python
sorted(zip(rf1.feature_importances_, data1.columns), reverse=True)
```
